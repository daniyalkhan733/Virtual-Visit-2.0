```r
# Daniyal Khan || 221A061

# Load necessary libraries
library(tm)

# Define text data as a character vector
text_data <- c(
  "The rapid advancements in artificial intelligence are reshaping industries.",
  "Machine learning algorithms are being used to solve complex problems.",
  "Data scientists are utilizing large datasets to uncover hidden patterns.",
  "Robust data preprocessing is critical for effective machine learning models.",
  "The development of AI technologies is accelerating innovation in healthcare.",
  "Automated decision-making processes are increasingly becoming part of business operations.",
  "Natural language processing enables machines to understand and generate human language.",
  "Visualization tools play a crucial role in presenting analytical results.",
  "Big data is transforming the way businesses interact with customers.",
  "The ethical implications of AI and machine learning need to be carefully considered."
)

# Convert text into a Corpus
corpus <- Corpus(VectorSource(text_data))

# Text cleaning: Convert to lower case, remove punctuation, numbers, and stopwords
corpus <- tm_map(corpus, content_transformer(tolower))    # Convert to lowercase
corpus <- tm_map(corpus, removePunctuation)               # Remove punctuation
corpus <- tm_map(corpus, removeNumbers)                   # Remove numbers
corpus <- tm_map(corpus, removeWords, stopwords("english"))  # Remove stopwords
corpus <- tm_map(corpus, stripWhitespace)                 # Remove extra whitespace

# Create a term-document matrix
tdm <- TermDocumentMatrix(corpus)

# Convert the term-document matrix to a matrix
m <- as.matrix(tdm)

# Get word frequency
word_freqs <- sort(rowSums(m), decreasing = TRUE)

# Convert to a dataframe
df <- data.frame(word = names(word_freqs), freq = word_freqs)

# Print the dataframe
print(head(df,10))

set.seed(1234)
```

## Creating a word cloud for visualization

```r
# Daniyal Khan || 221A061

set.seed(1234)
wordcloud(words = df$word,
          freq = df$freq,
          min.freq = 1,
          max.words = 100,
          random.order = TRUE,
          rot.per = 0.25,
          colors = brewer.pal(8, "Dark2"))
```
