## Importing Important Libraries

```python
import nltk
from nltk.corpus import twitter_samples
import matplotlib.pyplot as plt
import random
```

## Download Dataset of twitter

```python
nltk.download('twitter_samples')
```

## Calculating length of data set


```python
tweets = twitter_samples.strings();
len(tweets)
```

```python
import pandas as pd
all_tweets = pd.DataFrame(tweets, columns=["Tweets"])
all_tweets.head()
```

## check for files

```python
twitter_samples.fileids()
```

## Formatting the tweets into strings



```python
positive_tweets = twitter_samples.strings('positive_tweets.json')
negative_tweets = twitter_samples.strings('negative_tweets.json')
```

## printing tweets count

```python
print('Number of positive tweets: ', len(positive_tweets))
print('Number of negative tweets: ', len(negative_tweets))
```

## Check for positive tweets

```python
positive_tweets = pd.DataFrame(positive_tweets, columns=["Tweets"])
positive_tweets["Sentiments"] = "positive"

positive_tweets.head()
```

## check for negative tweets

```python
negative_tweets = pd.DataFrame(positive_tweets, columns=["Tweets"])
negative_tweets["Sentiments"] = "negative"

negative_tweets.head()
```

## Merging both

```python
tweets = pd.concat([positive_tweets, negative_tweets], ignore_index=True)
tweets = tweets.sample(frac=1).reset_index(drop=True)
tweets[1:15]
```

## Now Visualizing this

```python
# Assuming you want to visualize the sentiment distribution
sentiment_counts = tweets['Sentiments'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.title('Sentiment Distribution in Twitter Dataset')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.show()
```

## Preprocessing the data

Data preprocessing is one of the critical steps in any machine learning project. It includes cleaning and formatting the data before feeding into a machine learning algorithm. For NLP, the preprocessing steps are comprised of the following tasks:  

- Tokenizing the string  
- Lowercasing  
- Removing stop words and punctuation  
- Stemming  

## Tokenizing the string




```python
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import TweetTokenizer
```

```python
tweet = tweets["tweets"][12]
print(tweet)
```

```python
tweet1 = re.sub(r'^RT[\s]+','',tweet)

tweet1 = re.sub(r'https?://[^\s\n\r]+','', tweet1)

tweet1 = re.sub(r'#','',tweet1)
tweet1 = re.sub(r'@','', tweet1)
tweet1 = re.sub(r'[:)|:()]','', tweet1)

print('\nOriginal tweet: \n', tweet)
print('\nCleaned tweet: \n', tweet1)

```

```python
tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
tweet_tokens = tokenizer.tokenize(tweet1)

print("Cleaned tweet:/n ",tweet1)
print("\nTokenized tweet:/n ",tweet_tokens)
```
